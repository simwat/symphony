```python
# Calculates the shift in XRD peak for Mg composition in ZnO

import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np

def func(x, a, b, c, d):
    return a * (np.arctan(b*(x - c))) + d

x = [0, 3, 7, 14, 19, 25]
x = np.array(x)
c_axis = [5.202, 5.201, 5.201, 5.19, 5.182, 5.181]

fig = plt.figure()
ax = plt.subplot(111)
plt.plot(x, c_axis, 'ko')

popt, pcov = curve_fit(func, x, c_axis)
ax.plot(x, func(x, *popt), 'r-',
         label='fit: a=%5.3f, b=%5.3f, c=%5.3f, d=%5.3f' % tuple(popt))

plt.xlabel('% Mg')
plt.ylabel('c-axis (A)')
ax.legend(bbox_to_anchor=(1.1, 1.05))
plt.show()
```
