```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

#FileNames = [0]*1
#for i in range(len(FileNames)):
    #string = 1
    #FileNames[i] = [string]
    #print(FileNames[i])

FileNames = ['2020061901', '2020061903', '2020061903 (1)'] 

fig = go.Figure()

for j in range(len(FileNames)):

    df = pd.read_csv(FileNames[j])
    dataWheader = np.array(df)

    data = dataWheader[2::]
    Wavelength = [0]*len(data) 
    Count = [0]*len(data)
    MaxLoc = [0]*len(data)
    for i in range(len(data)):
        step_0 = str(data[i]).split('\\t')
        Wavelength[i] = float(step_0[0][2::])
        Count[i] = float(step_0[1][0:-2])
    
     # Add traces
    fig.add_trace(go.Scatter(x=Wavelength, y=Count,
                    mode='lines',
                    name='%.0f' %j))

#plt.legend()
#plt.xlim(364.85, 365.025)
#lgd = ax.legend(bbox_to_anchor=(1.1, 0.1))
#text = ax.text(-0.2,1.05, "Default25to35", transform=ax.transAxes) # CHANGE K_RANGE
#go.savefig('June18_recalibration2_cursor.png', bbox_extra_artists=(lgd, text), bbox_inches='tight')

fig['layout']['xaxis'].update(title='Wavelength (nm)', range=[366.6, 370], dtick=0.25, autorange=False)
fig['layout']['yaxis'].update(title='Counts/sec', autorange=True)
fig.show()
  
    
```
