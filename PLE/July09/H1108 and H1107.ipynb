```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

#E = hc/lambda in eV calculations
h = 6.62607004*10**(-34) # m^2kg/s
c = 299792458 # m/s
eV = 1.602176634*10**(-19) # J
Conversion = h*c/eV

FileNames = [0]*4
for i in range(len(FileNames)):
    if i <= 0:
        string = '202007090%.0f' %(i+3)
    elif i <= 2:
        string = '202007100%.0f' %(i+5)
    else:
        string = '2020070902'
    FileNames[i] = [string]
    print(FileNames[i])

fig = go.Figure()

names = ['H1108 July 9', 'H1108 July 10 ZnO scan', 'H1108 July 10 Mg scan', 'H1107']

for j in range(len(FileNames)):
    
    
    Wavelength, Count = np.genfromtxt(FileNames[j][0], delimiter="\t", unpack = True, skip_header=4)
    EnergyEV = [0]*len(Wavelength)
    
    Wavelength = Wavelength*10**(-9)*1.000289
    EnergyEV = Conversion/Wavelength*1000
    
    # Add traces
    fig.add_trace(go.Scatter(x=EnergyEV, y=Count,
                    mode='lines',
                    name=names[j]))
                            
fig['layout']['xaxis'].update(title='Energy (meV)', autorange=True)
fig['layout']['yaxis'].update(title='Counts/sec', autorange=True)
fig.update_layout(showlegend=True,  yaxis_type="log")
fig.show()
```
