```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

FileNames = ['Default', 'Default (1)', 'Default (2)', 'Default (3)',
            'Default (4)', 'Default (5)', 'Default (6)', 'Default (7)',
             'Default (8)','Default (9)', 'Default (10)', 'Default (11)', 
             'Default (12)', 
             'Default (13)', 'Default (14)', 'Default (15)']

fig = go.Figure()

for j in range(len(FileNames)):

    df = pd.read_csv(FileNames[j])
    dataWheader = np.array(df)

    data = dataWheader[2::]
    Wavelength = [0]*len(data) 
    Count = [0]*len(data)
    MaxLoc = [0]*len(data)
    for i in range(len(data)):
        step_0 = str(data[i]).split('\\t')
        Wavelength[i] = float(step_0[0][2::])
        Count[i] = float(step_0[1][0:-2])

    # Add traces
    fig.add_trace(go.Scatter(x=Wavelength, y=Count,
                    mode='lines',
                    name='%.0f' %j))

plt.legend()
plt.xlim(364.85, 365.025)
#lgd = ax.legend(bbox_to_anchor=(1.1, 0.1))
fig.show()
```
