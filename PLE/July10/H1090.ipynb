```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

#E = hc/lambda in eV calculations
h = 6.6261*10**(-34) # m^2kg/s
c = 2.9979*10**8 # m/s
eV = 1.6022*10**(-19) # J
Conversion = h*c/eV

FileNames = [0]*1
for i in range(len(FileNames)):
    if i <= 5:
        string = '202007100%.0f' %(i+2)
    else:
        string = '20200709%.0f' %(i+4)
    FileNames[i] = [string]
    print(FileNames[i])

fig = go.Figure()

for j in range(len(FileNames)):
    
    
    Wavelength, Count = np.genfromtxt(FileNames[j][0], delimiter="\t", unpack = True, skip_header=4)
    EnergyEV = [0]*len(Wavelength)
    
    Wavelength = Wavelength*10**(-9)*1.000289
    EnergyEV = Conversion/Wavelength*1000
    
    # Add traces
    fig.add_trace(go.Scatter(x=EnergyEV, y=Count,
                    mode='lines',
                    #name='%.0f' %j, 
                            name = 'H1090'))

fig['layout']['xaxis'].update(title='Energy (meV)', autorange=True)
fig['layout']['yaxis'].update(title='Counts/sec', autorange=True)
fig.update_layout(showlegend=True,  yaxis_type="log")
fig.show()
```
