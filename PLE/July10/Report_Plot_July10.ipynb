```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

#E = hc/lambda in eV calculations
h = 6.6261*10**(-34) # m^2kg/s
c = 2.9979*10**8 # m/s
eV = 1.6022*10**(-19) # J
Conversion = h*c/eV

FileNames = ['2020070908', '2020071002', '2020071005', '2020071007']

fig = go.Figure()

names = ['H1088', 'H1090', 'H1092', 'H1108 baseline']
offset = [0, 1.282, 1.099, -2.228]

for j in range(len(FileNames)):
    
    
    Wavelength, Count = np.genfromtxt(FileNames[j], delimiter="\t", unpack = True, skip_header=4)
    EnergyEV = [0]*len(Wavelength)
    
    Wavelength = Wavelength*10**(-9)*1.000289
    EnergyEV = Conversion/Wavelength*1000
    
    EnergyEV = [x - offset[j] for x in EnergyEV]
    
    # Add traces
    fig.add_trace(go.Scatter(x=EnergyEV, y=Count,
                    mode='lines',
                    #name='%.0f' %j, 
                            name = names[j]))

fig['layout']['xaxis'].update(title='Energy (meV)', dtick=5, autorange=True)
fig['layout']['yaxis'].update(title='Counts/sec', autorange=True)
fig.update_layout(showlegend=True)
fig.show()
```
