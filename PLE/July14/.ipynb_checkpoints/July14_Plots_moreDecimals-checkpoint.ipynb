```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

#E = hc/lambda in eV calculations
h = 6.62607004*10**(-34) # m^2kg/s
c = 299792458 # m/s
eV = 1.602176634*10**(-19) # J
Conversion = h*c/eV

FileNames = ['2020071402', '2020071404', '2020071406', '2020071408', '2020071410']

fig = go.Figure()

names = ['H1108 - baseline','H1107 - 2 hour baseline', 'H1088 - Mg', 'H1090 - Mg', 'H1092 - no Mg']

for j in range(len(FileNames)):
    
    
    Wavelength, Count = np.genfromtxt(FileNames[j], delimiter="\t", unpack = True, skip_header=4)
    EnergyEV = [0]*len(Wavelength)
    
    Wavelength = Wavelength*10**(-9)*1.000289
    EnergyEV = Conversion/Wavelength*1000
    
    # Add traces
    fig.add_trace(go.Scatter(x=EnergyEV, y=Count,
                    mode='lines',
                    #name='%.0f' %j, 
                            name = names[j]))

fig['layout']['xaxis'].update(title='Energy (meV)', dtick = 5, autorange=True)
fig['layout']['yaxis'].update(title='Counts/sec', autorange=True) # range = [1, 5.6], 
fig.update_layout(showlegend=True,  yaxis_type="log")
fig.show()
```
