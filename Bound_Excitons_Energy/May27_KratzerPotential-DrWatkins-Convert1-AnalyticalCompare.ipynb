Before, we worked on the Schrodinger equation for hydrogen:  
  
\begin{equation}
\frac{1}{r}\frac{d^2}{dr^2}(r\psi) = \frac{2m}{\hbar^2}(-E + V(r))\psi \\
\end{equation}

  
where V(r) = $- \frac{e^2}{r}$  
  
Then for excitons, the proposed Puls potential is:   
\begin{equation}
V(r) = \frac{e^2}{\epsilon_0r_h}[1 - 2e^\frac{-2r_h}{a_e}(1+\frac{r_h}{ae})] \\
\end{equation}
  
  
To compare with Matt Frick's work, here we will also use the Kratzer potential:
  
\begin{equation}
V(r_h) = -2D\left(\frac{b}{r_h} - \frac{b^2}{2r_h^2}\right)\\
\end{equation}

  
where:   
$D = \frac{se^2}{2\epsilon_0a_e}$  
$b = ta_e$
  
and:  
s = 1.01345  
t = 1.307
  
from Matt Frick's fit optimizations to get the Kratzer potential to  match the Puls Potential  
  
  
After the following assumptions and substitutions:  
  
  
\begin{equation}
\rho = \frac{r_h}{a_A}
\end{equation}
\begin{equation}
a_A = \frac{\hbar^2}{m_h e^2}
\end{equation}
\begin{equation}
Rydberg = \frac{m_he^4}{2\hbar^2}  
\end{equation}
\begin{equation}
\epsilon_0 = 1
\end{equation}

  
The Krazer potential becomes (in Rydberg units):
  
\begin{equation}
V(\rho_h) = -2s\left(\frac{t}{\rho_h} - \frac{t^2}{2\rho_h^2}\right)\\
\end{equation}  


Assuming that $a_e \ne a_A$ we have:

\begin{equation}
V(\rho_h) = -\frac{2s}{\rho_h}\left(t - \frac{t^2 a_e}{2\rho_h a_A}\right)\\
\end{equation}  
   
   
Symphony's continued result: in electron Rydberg units  

\begin{equation}
V(\rho_h) = -\frac{2s}{\rho_h}\left(t\left(\frac{m_h}{m_e}\right) - \frac{t^2 K}{2\rho_h}\left(\frac{m_h}{m_e}\right)^2\right)\\
\end{equation}  
  
With  
$m_h$ = 0.59  
$m_e$ = 0.28 
In atomic units  
K = 1.10  
For Matt Frick's Kratzer best agreement  
  
Lastly, to convert the electron Rydberg to effective Rydberg, then to meV:  
$Ry_e = \frac{m_e}{\epsilon^2}Ry = \frac{0.28}{8.5^2}(13.6~eV) = 0.0527~eV$ 
  
  
1 a.u. = 2 Ry  
  
OR: Like Matt Frick had, we can put $a_e$ = $\epsilon$/$m_e$

```python
import numpy as np

def MattKratzer(n, l):
    # calculate intermediate constants
    k = 1
    me = 0.28
    mh = 0.59
    ae = k*8.5/me
    b = 1.337/ae
    D = 1.0136/(2*8.5*ae)
    gamma = np.sqrt(2*mh*b**2*D)
    
    # calculate requested energy level
    E = -1/(2*mh*b**2)*gamma**4/(n + 0.5 + np.sqrt((l + 0.5)**2 + gamma**2))**2
    return E*13.6*2*1000
 
E_krat_Matt = [0]*9
count = 0
for n in range(3):
    for l in range(3):
        count += 1
        E_krat_Matt[count -1] = MattKratzer(n, l)

#converting to meV
E_krat_Matt = [x*13.6*2*1000 for x in E_krat_Matt]        
print(E_krat_Matt)
```
